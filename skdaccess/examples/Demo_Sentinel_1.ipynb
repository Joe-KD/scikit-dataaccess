```python
import matplotlib.pyplot as plt
plt.rcParams['figure.dpi'] = 150
import numpy as np
from getpass import getpass
```

```python
from skdaccess.geo.sentinel_1.cache import DataFetcher as S1DF
```

Supply Earth Data credentials

```python
username='Enter username'
```

```python
password = getpass()
```

Define urls for Sentinel 1 data and precise orbits

```python
slc_url_list = ['https://datapool.asf.alaska.edu/SLC/SA/S1A_IW_SLC__1SSV_20141103T195043_20141103T195057_003122_00395A_F396.zip']
satellite_url_list = ['https://s1qc.asf.alaska.edu/aux_poeorb/S1A_OPER_AUX_POEORB_OPOD_20141124T123237_V20141102T225944_20141104T005944.EOF']
```

Create data fetcher

```python
s1df = S1DF(slc_url_list, satellite_url_list, username, password, swath=3, polarization='VV')
```

Access data

```python
s1dw = s1df.output()
```

```python
label, data = next(s1dw.getIterator())
```

```python
plt.title(label, fontsize=8)
plt.imshow(np.abs(data[::10,::10]), vmin=0, vmax=100, cmap='gray', origin='lower')
plt.axis('off');
```
